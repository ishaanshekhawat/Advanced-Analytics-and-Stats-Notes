```python
from scipy.stats import binom
```

```python
# If it's given that probability that a student likes python is 66%, 
# find the probabilioty that 3 out of 5 students like python.

# Given values - 

n = 5
k = 3
p = 0.66
```

```python
Binomial_Probability = binom.pmf(k, n, p)
print(Binomial_Probability)
```

```python
# Hospital records show that of patients suffering from a specific disease, 75% die of it.
# What is the probability that of six randomly selected patients, 4 will recover?

n = 6
k = 4
p = 0.75

binom_probability = binom.pmf(k,n,p)
print(binom_probability)
```
