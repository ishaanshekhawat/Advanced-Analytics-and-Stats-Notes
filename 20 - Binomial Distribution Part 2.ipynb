```python
from scipy.stats import binom
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# If we picked 100 people randomly from this titanic data, 
# what is the probability distribution of the number of suviours?

titanic_df = pd.read_csv('titanic.csv')
```

```python
titanic_df['survival_binary'] = titanic_df['survived'].apply(lambda x : 1 if x==1 else 0)

print(titanic_df)

n = len(titanic_df)
k = titanic_df['survival_binary'].sum()
p = k/n
print(n,k,p)
```

```python
# Simulate 100 hypothetical passengers

trials = 100
x = np.arange(0,trials + 1) # from 0 to 100 suvivors
```

```python
# PMF & CDF
pmf_values = binom.pmf(x,trials,p)
cdf_values = binom.cdf(x,trials,p)
print(pmf_values)
print(cdf_values)
```

```python
plt.figure(figsize=(12,6))

# PMF plot ----------------------

plt.subplot(1,2,1)
plt.bar(x,pmf_values,color='pink',alpha=0.6)
plt.title('Titanic surival plot')
plt.xlabel("number of survival (out of 100)")
plt.ylabel("Probability")
plt.grid(axis='y',linestyle='--',alpha=0.6)


# CDF Plot

plt.subplot(1,2,2)
plt.plot(x,cdf_values,color='green',marker='o',linestyle='-')
plt.title('Titanic surival plot(100 hypothetical passangers)')
plt.xlabel("number of survival (out of 100)")
plt.ylabel("Cumulative Probability")
plt.grid()
plt.tight_layout()
plt.show()
```
