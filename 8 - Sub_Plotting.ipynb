```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tips.csv')  # loading the tips dataset from seaborn library
```

```python
# sub-plot 1 : Bar plot of average tip by day 
avg_tip_by_day = df.groupby('day')['tip'].mean()
plt.figure(figsize=(8,6))
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.bar(avg_tip_by_day.index, avg_tip_by_day.values, color=['orange','lightblue','pink'])
plt.title('Average Tip by Day')
plt.ylabel('Average Tip')
# sub-plot2 : pie chart of smoker vs non-smoker
smoker_counts = df['smoker'].value_counts()
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
plt.pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
plt.title('Smoker vs Non-Smoker Distribution')
plt.tight_layout()
plt.show()
```

```python
# Object-Oriented (OO) syntax
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 row, 2 columns
# sub-plot 1 : Line plot of total bill over time 20 rows
axes[0].plot(df['total_bill'].iloc[:20], color='blue', marker='o',linestyle='-')
axes[0].set_title('Total Bill Over Time (First 20 Rows)')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Total Bill')
# sub-plot 2 : Histogram bar of average tip by gender
avg_tip_by_gender = df.groupby('gender')['tip'].mean()
axes[1].bar(avg_tip_by_gender.index, avg_tip_by_gender.values, color=['lightpink', 'lightblue'])
axes[1].set_title('Average Tip by Gender')
axes[1].set_ylabel('Average Tip') 
plt.tight_layout()
plt.show()      
```
