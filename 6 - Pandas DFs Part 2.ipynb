```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("shopping.csv")
numeric_cols = ['CustomerID', 'Tenure_Months' , 'Transaction_ID' , 'Quantity','Avg_Price','Delivery_Charges','GST','Offline_Spend','Online_Spend','Month','Discount_pct']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    
```

```python
male_customers = df[df['Gender']=='M']
print(male_customers)
```

```python
male_chicago = df[(df['Gender'] == 'M') & (df['Location'] == 'Chicago')]
print(male_chicago)
```

```python
male_chicago_alt = df.query("Gender == 'M' and Location == 'Chicago'")
print(male_chicago_alt)
```

```python

categories = ['Nest-USA', 'Alexa-USA']
selected_products = df[df['Product_Category'].isin(categories)]
print(selected_products)
```

```python
categories = ['Nest-USA', 'Alexa-USA']
selected_products_alt = df.query("Product_Category in @categories")
print(selected_products_alt)
```

```python
multiple_qty = df[df['Quantity']>1]
print(multiple_qty) 
multiple_qty_alt = df.query("Quantity > 1")
print(multiple_qty_alt) 
```

```python
tenure_range = df[df['Tenure_Months'].between(6,12)]
print(tenure_range)
tenure_range_alt = df.query("6 < Tenure_Months <= 12")
print(tenure_range_alt)
```

```python
tenure_range = df[df['Tenure_Months'].between(6,12,inclusive='left')] # other options: 'both', 'neither', 'right'
print(tenure_range)
```

```python
tenure_range_alt = df.query("6 <= Tenure_Months <= 12")
```

```python
#filter rows with string methods
camera_products = df[df['Product_Description'].str.contains('Camera', case=False,na=False)]
print(camera_products)
```

```python
camera_Product_alt = df[df['Product_Description'].str.contains('Camera', case=False, regex=True, na=False)]
print(camera_Product_alt)   
```

```python
#filter rows based on null or non-null values
with_coupons = df[df['Coupon_Code'].notnull()]
print(with_coupons)
```

```python
with_coupons_alt = df.query("Coupon_Code == Coupon_Code")
print(with_coupons_alt)
```

```python
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
jan_2019 = df[(df['Transaction_Date'] >= '2019-01-01') & (df['Transaction_Date'] < '2019-02-01')]
print(jan_2019)
```

```python
jan_2019_alt = df.query("Transaction_Date >= '2019-01-01' and Transaction_Date < '2019-02-01'")
print(jan_2019_alt)
```

```python
# Extract specific columns after filtering
male_products = df[df['Gender'] == 'M'][['CustomerID', 'Product_Description', 'Quantity']]
print(male_products)
```

```python
male_products_alt = df.loc[df['Gender']=='M', ['CustomerID', 'Product_Description', 'Quantity']]
print(male_products_alt)
```

```python
# filter top N values usimg nlargest
top_online_spend = df.nlargest(5, 'Online_Spend')
print(top_online_spend) 
```

```python
top_online_spend_alt = df.sort_values(by='Online_Spend', ascending=False).head(5)
print(top_online_spend_alt) 
```

```python
# filtering string columns with startswith and endswith - safe for NANS
electronics_products = df[df['Product_Description'].str.startswith('Nest', na=False)]
print(electronics_products)
```

```python
electronics_products_alt = df[df['Product_Description'].str.endswith('USA', na=False)]
print(electronics_products_alt)
```

```python
# coupling based on filtering total spend per customer
customer_total_spend = df.groupby('CustomerID')['Online_Spend'].sum()
print(customer_total_spend)
```

```python
high_spenders = customer_total_spend[customer_total_spend > 5000]
print(high_spenders)
high_spenders_alt = df.groupby('CustomerID').filter(lambda x: x['Online_Spend'].sum() > 5000)
print(high_spenders_alt)
```
