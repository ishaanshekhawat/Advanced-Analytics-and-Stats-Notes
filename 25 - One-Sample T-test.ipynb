```python
import pandas as pd
from scipy import stats
```

```python
data = pd.read_csv('Mall_Customers.csv')
age_data = data['Age']
```

```python
age_data = age_data.sample(n = 30, replace = False)

# H0: Mall customers have an avreage age of 40
pop_avg_age = 40
```

```python
t_stat, p_value = stats.ttest_1samp(age_data, pop_avg_age)
print('P-Value: ', p_value)
```

```python
alpha = 0.05
if p_value > 0.05:
    print('Fail to reject H0')
else:
    print('Reject H0')
```
