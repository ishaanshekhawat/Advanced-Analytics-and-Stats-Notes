```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
data = pd.read_csv('tips.csv')

data.head()
```

```python
total_bill_series = data['total_bill']
```

```python
# Normality test with Shapiro-Wilk Test

shapiro_test = stats.shapiro(total_bill_series)
print(f'Test Statistic(W): {shapiro_test.statistic}')
print('p-value: ', shapiro_test.pvalue)
```

```python
alpha = 0.05 # Common Significance Level
```

```python
if shapiro_test.pvalue > 0.05:
    print('Data is normally distributed')
else:
    print('Data is not normally distributed')
```

```python
# QQ_Plot

stats.probplot(total_bill_series, dist="norm", plot=plt)
plt.title('QQPlot of Total Bill')
plt.xlabel('Theoritical Quantiles (Normal Distribution)')
plt.ylabel('Ordered Values (Total Bill)')
plt.grid(True)
plt.show()
```
