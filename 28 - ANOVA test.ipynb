```python
# One-way ANOVA test on Latest_mon_usage column 
# FOR ANOVA there is no need to sample, use the full dataset

from scipy.stats import f_oneway
import pandas as pd

df = pd.read_csv('cust_seg.csv')

df.head()
```

```python
print(df.segment.value_counts())
```

```python
s1 = df.Latest_mon_usage[df.segment == 1]
s2 = df.Latest_mon_usage[df.segment == 2]
s3 = df.Latest_mon_usage[df.segment == 3]
```

```python
f_stat, p_value = f_oneway(s1,s2,s3)

print('P-Value: ', p_value)
```

```python
if p_value < 0.05:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```
