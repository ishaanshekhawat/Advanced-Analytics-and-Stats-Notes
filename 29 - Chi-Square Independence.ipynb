```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('titanic.csv')

df.head()
```

```python
print(df.survived.value_counts())
print(df.pclass.value_counts())
```

```python
contingency_table = pd.crosstab(df['survived'], df['pclass'])

chi2_stat, p_value, degree_of_freedom, expected = stats.chi2_contingency(contingency_table)

print(chi2_stat)
print(degree_of_freedom)
print(p_value)
print(contingency_table)
print(expected)
```

```python
# Find if gender and smoking are related in the trips dataset

df1 = pd.read_csv('tips.csv')

df1.head()
```

```python
print(df1.gender.value_counts())
print(df1.smoker.value_counts())
```

```python
contingency_table = pd.crosstab(df1['gender'], df1['smoker'])

chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)
```

```python
print(chi2_stat)
print(dof)
print(p_value)
print(contingency_table)
print(expected)
```

```python
import numpy as np

df2 = pd.read_csv('diabetes.csv')
conditions = [(df2['Age'] <= 30), ((df2['Age'] > 30) & (df2['Age'] <= 50)), (df2['Age'] > 50)]
group = ['Young', 'Middle', 'Old']
df2['Age_group'] = np.select(conditions, group, default='Unknown')

df2.head()
```

```python
df2['Age_group'].value_counts()
df2['Outcome'].value_counts()
```

```python
contingency_table = pd.crosstab(df2['Age_group'], df2['Outcome'])

chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print(chi2_stat)
print(dof)
print(p_value)
print(contingency_table)
print(expected)
```
