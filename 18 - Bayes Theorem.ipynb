```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
total_passengers = len(df)
survived_count = len(df[df['survived']==1])
female_count = len(df[df['gender']=='female'])
female_survived_count = len(df[(df['gender']=='female') & (df['survived']==1)])
P_survived = survived_count/total_passengers
P_female = female_count/total_passengers
P_female_given_survived = female_survived_count/survived_count
```

```python
# Apply Bayes's Theorem
P_survived_given_female = P_female_given_survived*P_survived/P_female
print(P_survived_given_female)
```

```python
N = len(df1)
A = len(df1[df1['loan_status'] == 'Approved'])
B = len(df1[df1['cibil_score'] >= 700])

# P_A_given__B = ??

A_and_B = len(df1[(df1['loan_status'] == 'Approved') & (df1['cibil_score'] >= 700)])
```

```python
P_A = A/N    # Probability of Loan is approved
P_B = B/N    # Probability of Good Credit Score
P_A_and_B = A_and_B/N   # Probability that Loan is approved with Good Credit Score
P_B_given_A = P_A_and_B/P_A     # Probability that it was a Good Credit Score, given that Loan is approved
P_A_given__B = P_B_given_A*P_A/P_B    # Probability that loan will be approved, given that it was a good credit score  

P_A_given__B
```
