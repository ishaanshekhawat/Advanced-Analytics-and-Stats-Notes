```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# List of daily commuter times in minutes
commuter_times = [16, 8, 35, 17, 13, 15, 15, 5, 16, 25, 20, 20, 12, 10]

# Convert the list to a NumPy array for easier mathematical operations
data_arr = np.array(commuter_times)
```

```python
# -----------------------
# CALCULATE BOX-PLOT STATS
# -----------------------

# First quartile (25th percentile)
q1 = np.percentile(data_arr, 25)

# Third quartile (75th percentile)
q3 = np.percentile(data_arr, 75)

# Interquartile Range (IQR = Q3 - Q1)
iqr = q3 - q1

# Lower and upper fences for detecting outliers using the 1.5 * IQR rule
lower_fence = q1 - (1.5 * iqr)
upper_fence = q3 + (1.5 * iqr)

# Identify outliers: values below lower_fence or above upper_fence
outliers = data_arr[(data_arr < lower_fence) | (data_arr > upper_fence)]
print(outliers)

# Print summary statistics
print(f'Q1 : {q1}')
print(f'Q3 : {q3}')
print(f'IQR: {iqr}')
print(f'LOWER_FENCE:  {lower_fence}')
print(f'UPPER_FENCE:  {upper_fence}')
```

```python
# -----------------------
# VISUALIZE USING BOXPLOT
# -----------------------

plt.figure(figsize=(10, 5))

# Horizontal boxplot of the commuter times
sns.boxplot(data=data_arr, orient='h')

plt.title('Boxplot of Commuter Times')
plt.xlabel('Minutes')
plt.show()
```
