```python
# Understanding Monte Carlo Simulation
import numpy as np
```

```python
# Part 1
def roll_dice():
    # Simulation of rolling a dice twice, minimum number we get is 1, max is 6
    # Add the result of the two simulations, so possible values will be:
    # (1,1) or(1,2) ...(6,6)...Sum will be between 2 and 12
    # Run multiple times to verify
    return np.sum(np.random.randint(1,7,2)) # Generate an array of 2 random integers between 1 and 6 (both inclusive) and add them>
print(roll_dice())
```

```python
# Part 2
# Someone appraches us saying I will give you 5 dollars if you get 7 and take 1 dollar if you got other than 7
# How do we know what will happen?
# Our own "Monte Carlo Simulation" like function
def monte_carlo_simulation(runs = 1000):
    results = np.zeros(2) # An array, results[0] and result[1] initialized to two zeroes
    for _ in range(runs):
        if roll_dice() == 8:
            results[0] += 1
        else:
            results[1] += 1
    return results

# Test 2-3 times and calculate how much you will win versus lose
print(monte_carlo_simulation())
print(monte_carlo_simulation())
print(monte_carlo_simulation())
```

```python
# Part 3
# Now do it 1000 times 
# Takes some time
results_win = np.zeros(1000)
print("---------For winning-------------")
for i in range(1000):
    results_win[i] = monte_carlo_simulation()[0]

print(results_win)
results_lose = np.zeros(1000)
print("---------for Losing-------------")
for i in range(1000):
    results_lose[i] = monte_carlo_simulation()[1]
print(results_lose)
```

```python
# Let us plot it
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.hist(results_win, bins = 15)
plt.show()
```

```python
# Let us plot it
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.hist(results_lose, bins = 15)
plt.show()
```

```python
# Our win/loss
print(results_win.mean()) # General mean
print(results_win.mean()* 5) # What we will get as win on an average
print(1000 - results_win.mean()) # What we will pay on an average
print(results_win.mean()* 4.75) # Just a marginal change in win reward - see the impact
print(1000- results_win.mean()) # What we will pay on an average
print(results_win.mean()/1000) # Probability of the 'We will win' result
```
