```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
df = pd.read_csv('diabetes.csv')

df.head()
```

```python
glucose_series = df['Glucose']

shapiro_test = stats.shapiro(glucose_series)

print('Test Statistic (W): ', shapiro_test.statistic)
print('P-Value: ', shapiro_test.pvalue)
```

```python
if shapiro_test.pvalue > 0.05:
    print('Data is normally distributed')
else:
    print('Data is not normally distributed')
```

```python
# QQ_Plot

stats.probplot(glucose_series, dist="norm", plot=plt)
plt.show()
```

```python
df.count()
```

```python
sample_means = []
n_samples = 100
sample_size = 30

for i in range(n_samples):
    ith_sample = df['Glucose'].sample(sample_size, replace=True)
    sample_means.append(ith_sample.mean())

print(sample_means)
```

```python
# Distribution of Sample Means

plt.hist(sample_means, bins = 30, edgecolor = 'k', alpha=0.7)
plt.show()
```

```python
# QQ-Plot for Sample Mean

stats.probplot(sample_means, dist='norm', plot=plt)
plt.show()
```

```python
shapiro_test1 = stats.shapiro(sample_means)

print('Test Statistic (W): ', shapiro_test1.statistic)
print('P-Value: ', shapiro_test1.pvalue)
```
