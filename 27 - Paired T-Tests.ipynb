```python
# Paired t-test: A comapny has launched some offer.
# H0: Pre and post usage is the same (i.e., offer not effective).

from scipy.stats import ttest_rel
from math import sqrt # for calculating critical value
from scipy.stats import t
import pandas as pd
```

```python
df = pd.read_csv('cust_seg.csv')
df.head()
```

```python
usage_data = df[['pre_usage', 'Post_usage_1month']]
usage_data = usage_data.sample(n=30)

test_stat, p_value = ttest_rel(usage_data['pre_usage'], usage_data['Post_usage_1month'])

print('P-Value: ', p_value)
```

```python
if p_value < 0.05:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```
